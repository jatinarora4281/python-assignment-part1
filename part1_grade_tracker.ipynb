Task 1 — Data Parsing & Profile Cleaning

```python
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]
target_student_name = ""
print("Student Profile Cards")

for student in raw_students:
    name = student["name"].strip()
    name = name.title()

    roll_no = int(student["roll"])

    marks_str = student["marks_str"]
    marks_split = marks_str.split(", ")      # splits the marks into list
    marks_list = []
    for m in marks_split:
        marks_list.append(int(m))                 # converts marks_str in to int

    name_words = name.split()                # splitting of words of each name to check validity
    valid = True                             # validating name contains only alphabets
    for w in name_words:
        if not w.isalpha():
            valid = False

    print("="*35)                            #formatted profile cards
    print(f"Student  : {name} ({'✓ Valid name' if valid else '✗ Invalid name'})")
    print(f"Roll No. : {roll_no}")
    print(f"Marks    : {marks_list}")
    print("="*35)

    if roll_no == 103:
        target_student_name= name

print("Student with Roll no. 103")
print(target_student_name.upper())
print(target_student_name.lower())
```

Task 2 — Marks Analysis Using Loops & Conditionals

```python
student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]

print("Grading and Analysis of Marks for" ,student_name)
print("-"*50)
for i in range(len(subjects)):
    subject = subjects[i]
    mark = marks[i]

    if mark >=90:                 #grading system conditions
      grade = "A+"
    elif mark >=80:
      grade = "A"
    elif mark >=70:
      grade = "B"
    elif mark >=60:
      grade = "C"
    else:
      grade = "F"

    print(f"{subject} : {mark} ({grade})")
print("-"*50)
total = sum(marks)
print("Total Marks :" , total)

avg = total / len(marks)
print("Average Marks :" ,round(avg, 2))

max_marks = max(marks)
max_index = marks.index(max_marks)
print("Maximum Marks :" ,subjects[max_index], max_marks)

min_marks = min(marks)
min_index = marks.index(min_marks)
print("Minimum Marks :" ,subjects[min_index], min_marks)
```

```python
print("--- Add New Subjects ---")

new_count = 0                    # to count how many subjects are added

while True:
    subject = input("Enter subject name (type 'done' to stop): ")

    if subject.lower() == "done":        # stopping condition
        break

    marks_input = input("Enter marks (0-100): ")

    if marks_input.isdigit() == False:        # checking if input is number or not
        print("Please enter valid numeric marks.")
        continue                              # skip this iteration

    mark = int(marks_input)

    if mark < 0 or mark > 100:                # checking if marks are in valid range
        print("Marks should be between 0 and 100.")
        continue

    subjects.append(subject)                   # adding valid subject and marks
    marks.append(mark)

    new_count += 1                             # increasing count
print("-"*50)
print("SUMMARY")
print("Number of new subjects added:", new_count)     # count of new subjects added

total_marks = sum(marks)
updated_avg = total_marks / len(marks)

print("Updated average marks:", round(updated_avg, 2))   # new average marks
```

Task 3 — Class Performance Summary

```python
class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]
print("Class Performance Report")
print("-" * 40)
print(f"{'Name':<18} | {'Average':<7} | Status")
print("-" * 40)

#variables/containers for Class Summary
pass_count = 0
fail_count = 0
total_avg_sum = 0              # for class average
topper_name = ""
topper_avg = -1

for student in class_data:
    name = student[0]
    marks = student[1]

    avg = sum(marks) / len(marks)
    avg = round(avg, 2)

    if avg >= 60:                 # pass/fail status
        status = "Pass"
        pass_count += 1
    else:
        status = "Fail"
        fail_count += 1

    if avg > topper_avg:          # topper average marks and name idetification
        topper_avg = avg
        topper_name = name

    total_avg_sum += avg

    print(f"{name:<18} | {avg:>7} | {status}")
    print("-" * 40)
print("Class Summary")
print("-" * 40)
print("Passed:", pass_count)
print("Failed:", fail_count)

print("Topper:", topper_name, "-", topper_avg)

class_avg = total_avg_sum / len(class_data)
print("Class Average:", round(class_avg, 2))
```

Task 4 — String Manipulation Utility

```python
essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "

# Step 1: remove extra spaces (as required)
clean_essay = essay.strip()
print(f"Step 1 (Cleaned): {clean_essay}")

# converting to lowercase for consistent processing (extra step for safety)
clean_essay = clean_essay.lower()

# Step 2: convert to Title Case
print(f"Step 2 (Title Case): {clean_essay.title()}")

# Step 3: count occurrences of 'python'
python_count = clean_essay.count("python")
print(f"Step 3 (Count of 'python'): {python_count}")

# Step 4: replace 'python' with 'Python 🐍'
emoji_essay = clean_essay.replace("python", "Python 🐍")
print(f"Step 4 (After Replacement): {emoji_essay}")

# Step 5: split into sentences
sentences = clean_essay.split(". ")
print(f"Step 5 (Split List): {sentences}")

# Step 6: print numbered sentences properly
print("Step 6 (Numbered Sentences):")

for i, sentence in enumerate(sentences, 1):
    s = sentence.strip()

    # skip empty strings (just in case)
    if not s:
        continue

    # add full stop if missing
    if not s.endswith("."):
        s += "."

    # capitalize first letter for neat output
    print(f"{i}. {s.capitalize()}")
```
